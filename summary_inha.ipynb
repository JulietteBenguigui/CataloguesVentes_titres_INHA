```python
file = "tous_les_titres.csv"
```

```python
import pandas as pd

df = pd.read_csv(file, sep=";")

df.head(10)
```

```python
df = pd.DataFrame(df)
```

```python
from datasets import load_dataset, Dataset
import pandas as pd
from huggingface_hub import CommitInfo
from google.colab import userdata

userdata.get('huggingface')

titres = Dataset.from_pandas(df)
```

```python
titres = titres.train_test_split(test_size=0.2)
```

```python
titres["train"][0]
```

```python
from transformers import AutoTokenizer
from google.colab import userdata

userdata.get('huggingface')

checkpoint = "google-t5/t5-small"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

### Création d'une fonction "preprocess_function"
- On créer ici une fonction de "prétraitement"
- C'est le moment où l'on normalise, tokenize, gère les formats, prépare les données avant qu'elles soient nourries au modèle

```python
prefix = "summarize: "

def preprocess_function(examples):
    inputs = [prefix + doc for doc in examples["titre_long"]]
    model_inputs = tokenizer(inputs, max_length=None, truncation=True)

    labels = tokenizer(text_target=examples["titre_court"], max_length=None, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
tokenized_titres = titres.map(preprocess_function, batched=True)
```

```python
from transformers import DataCollatorForSeq2Seq
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=checkpoint)
```

```python
import evaluate
rouge = evaluate.load("rouge")
```

```python
import numpy as np

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    result = rouge.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)

    prediction_lens = [np.count_nonzero(pred != tokenizer.pad_token_id) for pred in predictions]
    result["gen_len"] = np.mean(prediction_lens)

    return {k: round(v, 4) for k, v in result.items()}
```

```python
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer

model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
```

```python
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer
import torch

training_args = Seq2SeqTrainingArguments(
    output_dir="INHA_in_short",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=4,
    predict_with_generate=True,
    fp16=True,
    push_to_hub=True,
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_titres["train"],
    eval_dataset=tokenized_titres["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)


trainer.train()
```

```python
trainer.push_to_hub()
```
